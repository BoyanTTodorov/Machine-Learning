### Pythagorean theorem, we are using to find the distance between two points in plane

$$ d = \sqrt{(x_2 - x_1) ^2 + (y_2 - y_1) ^2}

```python

import matplotlib.pyplot as plt
import numpy as np
import math
np.random.seed(0)

x = [np.random.randint(1, 10) for _ in range(2)]
y = [np.random.randint(1, 10) for _ in range(2)]

plt.style.use("ggplot")
plt.figure(figsize=(12, 6))
plt.grid(True)
plt.title("Pythagorean theorem distance between x, y", fontsize=14, fontweight="bold")
plt.xlabel("X", fontsize=14, fontweight="bold")
plt.ylabel("Y", fontsize=14, fontweight="bold")
plt.scatter(x, y, color="red")
plt.plot(x, y, linestyle="--", color="blue")
plt.xlim(0, 10)
plt.ylim(0, 5)
plt.text(x[0], y[0] - 0.5, f"({x[0]}, {y[0]})", fontsize=12, ha='center')
plt.text(x[1], y[1] - 0.5, f"({x[1]}, {y[1]})", fontsize=12, ha='center')
plt.text(abs(x[0] - x[1]) / 2, y[0] - 0.5, f" Distance : {int(math.sqrt((x[1] - x[0])**2 + (y[1] - y[0])**2))}", fontsize=12, color="blue")
plt.show()
```
